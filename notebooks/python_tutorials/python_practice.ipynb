```python
def greetings():
    return 'Welcome to Python'
print(greetings()) 

def uppercase_decorator(function):
    def wapper():
        func=function()
        return func.upper()
    return wapper

upper=uppercase_decorator(greetings)
print(upper())
```

```python
#same functionality
def uppercase_decorator(function):
    def wapper():
        func=function()
        return func.upper()
    return wapper

@uppercase_decorator
def greetings():
    return "Welcome to Python"

greetings() 
```

```python
#unpacking the dictionary
def unpacking_person_info(name, country, city, age):
    return f'{name} lives in {country}, {city}. He is {age} year old.'
dct = {'name':'Asabeneh', 'country':'Finland', 'city':'Helsinki', 'age':250}
print(unpacking_person_info(dct)) # Asabeneh lives in Finland, Helsinki. He is 250 years old.
```

```python
print(unpacking_person_info(**dct)) 
```

```python
#this is arbitary number of attributes comes in to picture
def sum_all(*args):
    print(f'sum of the values {sum(args)}')
    return args

sum_all(1,2,3,4,5)
```

```python
def unpacking_dictionary(**kwargs):
    for key in kwargs:
        print(f"{key} = {kwargs[key]}")
    return kwargs

unpacking_dictionary(name="Asabeneh",
      country="Finland", city="Helsinki", age=250)
```

```python
class Person:
    class_name= 'Its a class element'
    
    def __init__(self):
        self.obj_name="its a object element"
    

p1=Person()
p2=Person()

print(p1.obj_name, p1.class_name)
print(p2.obj_name, p2.class_name)
```

```python
# Lets change it
p1.class_name="Its a changed class name"
p1.obj_name="p1 class name"

print(p1.obj_name, p1.class_name)
print(p2.obj_name, p2.class_name)
```

```python
class PersonGender(Person):
    def __init__(self,name, gender):
        self.name=name
        self.gender=gender
        super().__init__()

    def get_personinfo(self):
        print(f'Hi,My name is {self.name} and I am {self.gender}')

p3=PersonGender('Chandra', 'Male') 

p3.get_personinfo()
```
